```python
import bs4
import requests
```

```python
URL = "https://codeforces.com/contest/1922/standings/page/1"
data = requests.get(URL)
soup = bs4.BeautifulSoup(data.text, 'html.parser')
```

```python
# Get class datatable
datatable = soup.select_one(".datatable")
rows = datatable.select("tr")

TOTAL_PROBLEM = 6
USER_ID = 0

[
    [{'is_passed','total_attempted','time_used'}]
]

for row in rows:
    tds = row.select("td")

    instance_data = []
    finished_minute = []
    
    problem_order = 0
    for td in tds:

        is_passed = 0
        total_attempted = 0
        time_used = -1

        if td.select("span.cell-accepted") and td.select("span.cell-time"):
            attempted_until_passed = td.select_one("span.cell-time").text
            time_used = td.select_one("span.cell-time").text

            if '+' in attempted_until_passed:
                is_passed = 1
            elif '-' in attempted_until_passed:
                is_passed = 0
            
            if attempted_until_passed == '+':
                total_attempted = 1
            else:
                total_attempted = int(attempted_until_passed[1])
            

            finished_minute.append([problem_order, time_used])
            

        elif td.select("span.cell-rejected"):
            pass
```
